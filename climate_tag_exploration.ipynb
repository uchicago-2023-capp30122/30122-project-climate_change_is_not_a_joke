Strategy 2:

Our next strategy was to tag projects based off of climate tokens. 

We did this by combining the known climate data frames and creating tokens from
the project names. We then found what tokens occured most frquently across all 
projects. From there we then limited the list for words related to climate. 

Next we created a function that then used this list to classify a single project.

```python

import spacy.cli
from project_tracker.load_and_clean import Cleaner


def add_climate_tag(df):

    """
    Creates new column and determines if project is climate realted based off
    of list of Tag words 

    Input:
        df(pandas data frame): data frame

    Return:
        data frame 
    """

    sp = spacy.load("en_core_web_lg")
    tag = pd.read_csv('data/raw/adb_19-21_climate_data/climate_tag_words.csv')

    print(tag)
    
    for index, row in df.iterrows():

        doc = sp(row['Project Name'])
        words = [token.text for token in doc if not token.is_stop and not token.is_punct and not token.like_num]
        for word in words:
            if word == '2)-' or word == '1)-':
                continue
            if tag["Tag"].str.contains(word).any():
                df.loc[index,['Climate-Related']] = 'Yes'
                break
            else:
                df.loc[index,['Climate-Related']] = 'No'   
    return df 


climate_tag = Cleaner.merg_climate_df()
climate_tag = add_climate_tag(climate_tag)
```

```python
climate_tag['Climate-Related'].value_counts(1)
```

Conclusion: 

We were able to capture over 60 percent of the known climate projects with the

because of this consistency we belive this approach was ultimately more 
accurate.
